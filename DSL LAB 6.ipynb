<pre>

Like Merge Sort, QuickSort is a Divide and Conquer algorithm. It picks an element as pivot and partitions the given array around the picked pivot. There are many different versions of quickSort that pick pivot in different ways.
 
 
1. Always pick first element as pivot.
2. Always pick last element as pivot (implemented below)
3. Pick a random element as pivot.
4. Pick median as pivot.
 
</pre>

<div>

Basically it just: <br> 
1) Selects the Last Element as the <b> Pivot </b> <br>
2) Places the <b> Elements smaller than the pivot to the right of the Pivot </b> <br>
3) Places the <b> Elements larger than the pivot to the left of the Pivot </b> <br>
4) Continues performing steps 1,2,3 recursively untill all numbers are not sorted <br>
    
</div>

```python
#print("QUICKSORT")


def partition(arr, low, high):
    print(" ")
    print("Partition")
    print("Low : ", low, " | ", "High : ", high)
    # index of smaller element
    #print("\ni = index of smaller element, set to [low - 1] = -1")
    i = (low - 1)
    print("\ni = index of smaller element, set to [low(", low, ") - 1] =", i)
    print("Partition at i :", i, ", i.e.", arr[i])
    # pivot
    pivot = arr[high]
    print("PIVOT : ", pivot)

    print("\nj is the Looping Variable")
    for j in range(low, high):
        # print(" ")
        # print("Low : ", low, "   |  ", "High : ", high)

        # If current element is smaller than or equal to pivot
        print("\nChecking if current element is smaller than or equal to pivot")
        #print("Current Element (j) : ", arr[j]," <= ",  "Pivot : ", pivot)
        print("Current Element(j)"," | ",  "Pivot")
        print("     ", arr[j], "        <=     ",pivot)

        if arr[j] <= pivot:
            print(" ")
            print("arr[j]", "<=", "Pivot")
            print(arr[j], " <=", pivot)
            # print("arr[j] : ", arr[j], "<=", "Pivot : ", pivot)
            # increment index of smaller element
            print("\nIncrement index of smaller element")
            print("i = ", i, "+", 1, ":")
            i = i + 1
            print("i = ", i)
            
            print("\nHere we are swapping '", arr[i], "' & '", arr[j], "'")
            print("Before Swapping : ")
            print("i : ", i, " | ", "arr[i] : ", arr[i])
            print("j : ", j, " | ",  "arr[j] : ", arr[j])
            
            arr[i], arr[j] = arr[j], arr[i]
            
            print(" ")
            print("After Swapping")
            print("i : ", i, " | ", "arr[i] : ", arr[i])
            print("j : ", j, " | ",  "arr[j] : ", arr[j])
            print("\nAfter Swapping Array : ", arr)
            print("\nIF LOOP OVER")

    print(" ")
    print("\ni & j have crossed each other so setting new pivot")
    print("For this we have to Swap arr[i + 1], arr[high]")
    print("i.e., swap (", arr[i + 1], ") & (", arr[high], ")")
    
    print("\nBefore Swapping (", arr[i + 1], ") & (", arr[high], ") :", arr)
    
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    
    print("\nAfter Swapping (", arr[i + 1], ") & (", arr[high], ") :", arr)
    
    print("\ni+1  : ", i+1, " | ", "arr[i+1]  : ", arr[i + 1])  
    print("High : ", high, " | ", "arr[high] : ", arr[high], "(New Pivot)")
    print(" ")
    print("ARRAY : ", arr)
    #print("end i+1 : ", i + 1)
    return (i + 1)


def quickSort(arr, low, high):
    print(" ")
    print("Low : ", low, " | ", "High : ", high)
    if len(arr) == 1:
        print("Lenght of array == 1 : ", arr)
        return arr
    if low < high:
        print("Low", "<", "High")
        print(" ", low, "<", high)
        
        # pi is partitioning index, arr[p] is now at right place
        #print("pi is partitioning index, arr[p] is now at right place")
        pi = partition(arr, low, high)
        print("\nPi : ", pi)
        
        # Separately sort elements before partition and after partition
        print("\nSeparately sort elements before partition and after partition")
            
        print("\nQuickSort\n")
        
        print("\nThis is to sort the elements to the left of Pivot")
        quickSort(arr, low, pi - 1)
        
        print("\nThis is to sort the elements to the right of Pivot")
        quickSort(arr, pi + 1, high)


# Driver code to test above
arr = [100, 12.51, 72.1, 85, 98.5, 35.0, 56.54]
# arr = [100, 72, 85, 98, 35, 56]
n = len(arr)

marks = ""
print(" ")
for j in range(n):
    marks = marks + ("| %s" % arr[j] + "% ")
print("Given Percentages  : ", marks)
print(" ")

quickSort(arr, 0, n - 1)

sorted_marks = ""
for i in range(n):
    # sorted_marks = sorted_marks + ("| %d" % arr[i] + "% ")
    sorted_marks = sorted_marks + ("| %s" % arr[i] + "% ")
print(" ")
print("Sorted Percentages : ", sorted_marks)


print(" ")
print("Top Ranks")

for l in range(1, 6):
    # print(l, " | ", "%d" % arr[-l], "%")
    print(l, " | ", "%s" % arr[-l], "%")
```
